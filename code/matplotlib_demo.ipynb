```python
from matplotlib import pyplot as plt
from matplotlib import style
import numpy as np
```

```python
# 1 plot graph

x=[1,7,4,5,6]
y=[5,8,9,2,4]
plt.plot(x,y)
plt.show()
```

```python
# 2 add title, xlabel, ylabel

x=[1,2,3,4,5]
y1=[34,33,56,78,40]
y2=[3,7,9,3,5]
plt.plot(x,y1)
plt.plot(x,y2)
plt.title('graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# 3 change to gridline, width of line, color, line labels 

style.use('dark_background')

x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,4,6,8,10]
plt.plot(x,y1,label='square',color='green',linewidth=15)
plt.plot(x,y2,label='double',color='white')
plt.title('graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(True,color='red')
plt.show()
```

```python
# 4 bar graph

x=[1,7,4,5,6]
y=[5,8,9,2,4]

plt.bar(x,y)
plt.show()
```

```python
# 5 difference between bar graph and histogram
# 6 histogram
# 7 histogram histtype, rwidth

ages=[22,33,22,40,90,10,5]
bins=[0,10,20,30,40,50,60]
plt.hist(ages,bins,histtype='stepfilled',rwidth=0.8)
plt.title('Histogram')
plt.show()
```

```python
# 8 legend()
# 9 scatter plot


x=[1,7,24,5,6,17,8,9,10,4,5]
y=[5,8,19,2,24,34,7,25,1,11,15]

x1=[1,7,4,5,16,17,8,9,10,4,15]
y1=[15,8,19,12,24,14,7,25,1,11,15]
plt.scatter(x,y)#,z)
plt.scatter(x1,y1)
plt.show()
```

```python
# 10 stack/area plot

day=[1,2,3,4,5]
sleep=[2,3,4,3,2]
eat=[7,8,7,2,2]
work=[2,3,4,3,2]
play=[8,5,7,8,13]

plt.plot([],[],color='m',label='day',linewidth=5)
plt.plot([],[],color='c',label='sleep',linewidth=5)
plt.plot([],[],color='r',label='eat',linewidth=5)
plt.plot([],[],color='k',label='work',linewidth=5)
plt.plot([],[],color='w',label='play',linewidth=5)

plt.stackplot(day,sleep,eat,work,colors=['m','c','r','k','w'])
plt.title('STACK / AREA ')
plt.legend()
plt.show()
```

```python
# 11 pie chart
# 12 pie chart : explode,autopct

slices=[7,2,2,13]
activities=['sleep','eat','work','play']
color=['c','m','r','b']

plt.pie(slices,labels=activities,colors=color,startangle=90,shadow=True,explode=(0,0,0,0.5),autopct='%1.1f%%')

plt.show()
```

```python
# 13 working with multiple plots

def f(t):
    return np.exp(-t) * np.cos(2* np.pi *t)

t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)

plt.subplot(221)
plt.plot(t1,f(t1),'bo',t2,f(t2))

plt.subplot(222)
plt.plot(t2,np.cos(2* np.pi *t2))#,'bo',t2,f(t2))
plt.show()
```
