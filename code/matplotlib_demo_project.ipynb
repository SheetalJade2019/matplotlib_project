```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style
import matplotlib.markers


company_sales=pd.read_csv('E:\\Python\\dataset\\company_sales_data.csv')
print(company_sales.head(3))
```

```python
# 1 Read Total profit of all months and show it using a line plot
# 2 Get Total profit of all months and show line plot with the following Style properties
'''
Style properties :

Line Style dotted and Line-color should be red
Show legend at the lower right location.
X label name = Month Number
Y label name = Sold units number
Add a circle marker.
Line marker color as read
Line width should be 3
'''
x=company_sales['month_number']
y=company_sales['total_profit']

plt.plot(x,y,color='red',linewidth=3,linestyle='dotted',marker='o')
plt.xlabel('Month Number')
plt.ylabel('total profit')
plt.title('sold units number')
plt.legend(loc='lower right')
plt.show()

#print(matplotlib.style.available)
```

```python
# 3 Read all product sales data and show it  using a multiline plot

x=company_sales['month_number']
facecream = company_sales['facecream']
facewash  = company_sales['facewash']
toothpaste = company_sales['toothpaste']
bathingsoap = company_sales['bathingsoap']
shampoo = company_sales['shampoo']

fig, ax = plt.subplots()
ax.plot(x,facecream,marker='o',label='facecream')
ax.plot(x,facewash,marker='o',label='facewash')
ax.plot(x,toothpaste,marker='o',label='toothpaste')
ax.plot(x,bathingsoap,marker='o',label='bathingsoap')
ax.plot(x,shampoo,marker='o',label='shampoo')

ax.legend()

plt.show()
```

```python
# 4 Read toothpaste sales data of each month and show it using a scatter plot

x=company_sales['month_number'].tolist()
y=company_sales['toothpaste'].tolist()

plt.scatter(x,y,label='Tooth paste sales data')
plt.grid(linestyle='dashed')
plt.xlabel('month number')
plt.ylabel('number of units sold')
plt.legend()
plt.show()
```

```python
# 5 Read face cream and facewash product sales data and show it using the bar chart
import numpy as np
x=company_sales['month_number']
y1=company_sales['facecream']
y2=company_sales['facewash']

bar_width=0.25

# set bar position on x axis
# plot bar
plt.bar(x,y1,width=bar_width,label='facecream')
plt.bar(x+bar_width,y2,width=bar_width,label='facewash')

plt.legend()
plt.show()
```

```python
# 6 Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk

x=company_sales['month_number'].tolist()
y=company_sales['bathingsoap'].tolist()

plt.bar(x,y,label='bathingsoap')
plt.xlabel('month number')
plt.ylabel('number of units sold')
plt.legend()
plt.savefig('save_bar.png')
plt.show()
```

```python
# 7 Read the total profit of each month and show it using the histogram to see most common profit ranges

profit=company_sales['total_profit'].tolist()
profit_range=[150000,200000,250000,300000,350000,400000,450000]
plt.hist(profit,profit_range,label='profit data')
plt.ylabel('profit')
plt.xlabel('profit range')
plt.show()
```

```python
# 8 Calculate total sale data for last year for each product and show it using a Pie chart

month=company_sales['month_number'].tolist()
labels=['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']

#product=[
#            company_sales['facecream'].sum(), company_sales['facewash'].sum(), company_sales['toothpaste'].sum(), 
#            company_sales['bathingsoap'].sum(), company_sales['shampoo'].sum(), company_sales['moisturizer'].sum()
#        ]
product=[facecream, facewash, toothpaste, bathingsoap, shampoo, moisturizer]
color=['b','g','r','m','c','k']
#plt.axis('equal')
plt.pie(product ,labels=labels, colors=color)
#print(len(labels))
plt.legend()
plt.show()
```

```python
# 9 Read Bathing soap facewash of all months and display it using the Subplot

x=company_sales['month_number'].tolist()
y1=company_sales['bathingsoap'].tolist()
y2=company_sales['facewash'].tolist()

plt.subplot(212)
plt.plot(x,y1,label='bathingsoap')

plt.subplot(211)
plt.plot(x,y2,label='facewash')

plt.xlabel('month')
plt.ylabel('sale')
plt.title('subplot of bathing soap and facewash')
plt.legend()
plt.show()
```

```python
# 10 Read all product sales data and show it using the stack plot

month=company_sales['month_number'].tolist()
facecream = company_sales['facecream']
facewash = company_sales['facewash']
toothpaste =  company_sales['toothpaste']
bathingsoap =  company_sales['bathingsoap']
shampoo = company_sales['shampoo']
moisturizer = company_sales['moisturizer']

plt.plot([],[],color='b',label='facecream')
plt.plot([],[],color='g',label='facewash')
plt.plot([],[],color='r',label='toothpaste')
plt.plot([],[],color='m',label='bathingsoap')
plt.plot([],[],color='c',label='shampoo')
plt.plot([],[],color='k',label='moisturizer')

plt.stackplot(month,facecream, facewash, toothpaste, bathingsoap, shampoo, moisturizer,colors=['b','g','r','m','c','k'])
plt.xlabel('month')
plt.ylabel('sale of products')
plt.title('all product sales data')
plt.legend()
plt.show()
```
